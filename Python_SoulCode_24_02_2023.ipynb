```python
# O operador booleano avalia expressões (expressão1 and expressão2 and expressão3...)
# Nesta lição vamos entender o operador bololeano "and".
# Todos tem que ser verdade para ser verdade
# Se um for falso, tudo é falso

n1 = 10
n2 = 5
n3 = 7

print(n1 > 5 and n2 == 5) # True and True = True

print(n1 > 1 and n3 < 5) # True and False = False

print(n1 !=10 and n3 >= 6) # False and True = False

print(n1 !=10 and n2 == n3) # False and False = False
```

```python
# Nesta lição vamos entender o operador bololeano "or".
# Basta um ser verdade, para todas as expressões serem verdade 

n1 = 10
n2 = 5
n3 = 7

print(n1 > 5 or n2 == 5) # True or True = True

print(n1 > 1 or n3 < 5) # True or False = True

print(n1 !=10 or n3 >= 6) # False or True = True

print(n1 !=10 or n2 == n3) # False or False = False
```

```python
# 1) Crie um algoritmo que leia um número e diga se ele está no intervalo fechado entre 25 e 200.

'''
1) Ler um número int (número)
2) Se o número for maior ou igual a 25 e menor igual a 200
'''

numero = int(input('Digite o primeiro número: '))

if numero >=25 and numero <= 200:
  print(f'O número {numero} está no intervalo fechado entre 25 e 200')

else:
  print(f'O número {numero} não está no intervalo fechado 25 e 200')

# if 25 <= numero <= 200: No python também temos essa notação
```

```python
# 2) Crie um algoritmo que leia 5 números e informe na tela o maior número lido.

'''
1) Ler 5 números inteiros (n1, n2, n3, n4, n5)
2) O primeiro número lido (n1) eu vou colocar em uma variável chamada (maior)
3) A partir do segundo número lido (n2) eu vou compará-lo com a variável (maior)
    Se o número lido no momento for maior que a variável (maior), então a variável maior recebe o número lido
'''

n1 = int(input("Digite o primeiro número: "))
maior = n1

n2 = int(input("Digite o segundo número: "))
if n2 > maior:
    maior = n2

n3 = int(input("Digite o terceiro número: "))
if n3 > maior:
    maior = n3

n4 = int(input("Digite o quarto número: "))
if n4 > maior:
    maior = n4

n5 = int(input("Digite o quinto número: "))
if n5 > maior:
    maior = n5

print("O maior número digitado foi:", maior)
```

```python
# 3) Crie um algoritmo que leia um número e informe na tela se o mesmo é positivo. Caso seja, mostre o número na tela

'''
1) Ler um número inteiro (n)
2) Se o número for maior que 0, informo que ele é positivo e mostro o resultado na tela
'''

n = int(input("Digite um número: "))

if n > 0:
    print(f'O número digitado foi {n} e ele é positivo')
```

```python
# 4) Crie um algoritmo que leia um número e informe se ele é divisível por 7.

'''
1) Ler um número inteiro (número)
2) Se o (número) divido por 7 resultar em resto 0, então ele é divisivel por 7
'''

numero = int(input("Digite um número inteiro: "))

if numero % 7 == 0:
    print(f'O número {numero} é divisível por 7')
else:
    print(f'O número {numero} não é divisível por 7')
```

```python
# 5) Crie um algoritmo que leia um número e diga se ele é par ou ímpar.

'''
1) Ler um número inteiro (n)
2) 
'''

n = int(input("Digite um número inteiro: "))

if n % 2 == 0:
    print(f'O número {n} é par')
else:
    print(f'O número {n} é ímpar!')
```

```python
"""
6) A prefeitura de Contagem abriu uma linha de crédito para os funcionários estatutários.
O valor máximo da prestação não poderá ultrapassar 30% do salário bruto. Fazer um
algoritmo que permita entrar com o salário bruto e o valor da prestação, e informar se
o empréstimo pode ou não ser concedido.
"""

# 1) Ler o valor do salario bruto (salario)
# 2) Ler o valor da prestação (prest)
# 3) Calcular 30% do salario bruto
# 4) Verificar se o valor da prestação é menor ou igual do que o valor do passo 3
#   Caso seja, posso conceder o empréstimo. Caso não sejam não posso conceder  


salario = float(input("Digite o valor do salário bruto: "))
prest = float(input("Digite o valor da prestação do empréstimo: "))

max_prest = salario * 0.3

if prest <= max_prest:
    print(f'Empréstimo pode ser concedido!')
else:
    print(f' O valor da prestação {prest} ultrapassa o valor máximo que é de {max_prest}! Empréstimo não concedido')
```
