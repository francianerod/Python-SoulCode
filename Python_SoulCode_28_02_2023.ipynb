## **Atividade 4**

Faça um programa que leia N notas e mostre na tela a média das notas lidas.

```python
# Entrada de dados
n = int(input('Digite a quantidade de notas que você vai digitar: '))
while n < 1:
  n = int(input('Digite a quantidade de notas que você vai digitar: '))

# Passo 1: Ler e somar a nota

soma = 0
contador = 1

while contador <= n:
  nota = float(input('Digitar uma nota de 0 a 10: '))
  while nota < 0 or nota > 10:
    nota = float(input('Nota inválida! Digite uma nota de 0 a 10: '))
  soma = soma + nota
  contador += 1


# Passo 2: Calculo da média
media = soma / n

print(f'A média das notas lidas foi de {media:.2f}')


# Somar as notas
# dividir pelo total de notas (n)
```

## **Atividade 8**

Desenvolver um algoritmo que leia um número não determinado de valores e calcule e escreva a média aritmética dos valores lidos, a quantidade de valores positivos, a quantidade de valores negativos e o percentual de valores negativos e positivos.


```python
soma_numeros = 0
qtd_numeros = 0
qtd_positivos = 0
qtd_negativos = 0 

while True: 
  numero = float(input('Digite um número qualquer: '))
  qtd_numeros += 1
  soma_numeros += numero
  if numero > 0:
    qtd_positivos += 1
  if numero < 0:
    qtd_negativos += 1
  
  cond = input('Digite n para sair ou qualquer outra coisa para continuar')
  if cond.lower() == 'n':
    break

media = soma_numeros / qtd_numeros

print(f'A media dos numeros digitados foi {media}')
print(f'A qtd de números positivos digitados foi {qtd_positivos}')
print(f'A qtd de números negativos digitados foi {qtd_negativos}')
print(f'O percentual de números positivos foi {qtd_positivos / qtd_numeros * 100}%')
print(f'O percentual de números negativos foi {qtd_negativos / qtd_numeros * 100}%')
```

## **Atividade 9**

Escrever um algoritmo que leia uma quantidade desconhecida de números e conte quantos deles estão nos seguintes intervalos: [0-25], [26-50], [51-75] e [76-100]. A entrada de dados deve terminar quando for lido um número negativo.

```python
intervalo1 = 0
intervalo2 = 0
intervalo3 = 0
intervalo4 = 0

while True:
  n = int(input('Digite um número qualquer: '))
  if n < 0:
    print('Obrigado por utilizar nosso sistema, Volte Sempre!')
    break
  
  if 0 <= n <= 25:
    intervalo1 += 1
  elif 26 <= n <= 50:
    intervalo2 += 1
  elif 51 <= n <= 75:
    intervalo3 += 1
  elif 76 <= n <= 100:
    intervalo4 += 1

print(f'A quantidade de valores no intervalo [0-25] foi de {intervalo1} número(s)')
print(f'A quantidade de valores no intervalo [26-50] foi de {intervalo2} número(s)')
print(f'A quantidade de valores no intervalo [51-75] foi de {intervalo3} número(s)')
print(f'A quantidade de valores no intervalo [76-100] foi de {intervalo4} número(s)')
```

## **Atividade 10**

Escreva um algoritmo que leia um valor inicial A e imprima a seqüência de valores do cálculo de A! e o seu resultado. Ex: 5! = 5 X 4 X 3 X 2 X 1 = 120

Sem título.png

```python
fatorial = 1

n = int(input('Digite um valor para o cálculo do fatorial: '))

while n < 0:
  n = int(input('Digite um valor para o cálculo do fatorial: '))

contador = n
while contador >= 1:
  if contador > 1:
    print(f'{contador} x', end=' ')
  
  fatorial *= contador
  contador -= 1

print(f'= {fatorial}')
```
